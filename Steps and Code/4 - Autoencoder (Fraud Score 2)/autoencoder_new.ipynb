# Autoencoder

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import os
os.environ['KERAS_BACKEND']='tensorflow'
from scipy import stats
import tensorflow as tf
import seaborn as sns
from pylab import rcParams
from sklearn.model_selection import train_test_split
from keras.models import Model, load_model
from keras.layers import Input, Dense
from keras.callbacks import ModelCheckpoint, TensorBoard
from keras import regularizers
```

```python
## Autoencoder
X_train, X_test = train_test_split(finalDF, test_size=0)

X_train = X_train.values
X_test = X_test.values
X_train.shape



input_dim = finalDF.shape[1]
encoding_dim = 4

input_layer = Input(shape=(input_dim, ))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
encoder_output= Dense(encoding_dim,)(encoder)


decoder = Dense(4, activation='relu')(encoder_output)
decoder = Dense(7, activation='tanh')(decoder)

autoencoder = Model(input=input_layer, output=decoder)

encoder= Model(input=input_layer, output=encoder_output)

autoencoder.compile(optimizer='adam', 
                    loss='mse')

autoencoder.fit(X_train, X_train,
                    epochs=15,
                    batch_size=64,
                    shuffle=True)

encoder.predict(X_train)
```

```python
predictions = autoencoder.predict(X_train)
DF1 = pd.DataFrame(data = predictions)
```
