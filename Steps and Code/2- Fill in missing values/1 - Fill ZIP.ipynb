```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv('NY property data.csv')
data.head()
```

```python
data.info()
```

## Fill in Field ZIP

```python
data[data['ZIP'].isnull()].head()
```

### Fill in missing values by B and BLOCK 

```python
# select record which has null in ZIP
zip = data[data['ZIP'].isnull()][['B','BLOCK','STADDR','ZIP']]
zip['RECORD'] = zip.index + 1 
zip.head()
```

```python
# select record with value in ZIP
data_withzip = data[data['ZIP'].notnull()]
data_withzip.head()
```

```python
# group record with value in ZIP by B and BLOCK
zipfill = data_withzip.groupby(['B','BLOCK'])[['ZIP']].agg(lambda x: stats.mode(x)[0][0]).reset_index()
zipfill.head()
```

```python
# Fill in the missing value using the most common ZIP value of records with same B and BLOCK
zip1 = zip.merge(zipfill, on = ['B','BLOCK'], how = 'left')
print(sum(zip1['ZIP_y'].isnull()))
print(sum(zip1['ZIP_y'].notnull()))
```

```python
zip1.head()
```

```python
zip1 = zip1[['RECORD','ZIP_y']]
zip1.head()
```

```python
## left join the original table and table with filled value by RECORD
data = data.merge(zip1, on = ['RECORD'], how = 'left')
data.head()
```

```python
data.ZIP = data.ZIP.fillna(data['ZIP_y'])
del data['ZIP_y']
data.head()
```

```python
data[['ZIP']].info()
```

### Filling in missing values by B and STADDR group 

```python
# select record which has null in ZIP
zip = data[data['ZIP'].isnull()][['B','BLOCK','STADDR','ZIP']]
zip['RECORD'] = zip.index + 1 
zip.head()
```

```python
zipfill2 = data_withzip.groupby(['B','STADDR'])[['ZIP']].agg(lambda x: stats.mode(x)[0][0]).reset_index()
zipfill2.head()
```

```python
zip2 = zip.merge(zipfill2, on = ['B','STADDR'], how = 'left')
del zip2['ZIP_x']
print(sum(zip2['ZIP_y'].isnull()))
print(sum(zip2['ZIP_y'].notnull()))
zip2.head()
```

```python
zip2 = zip2[['RECORD','ZIP_y']]
zip2.head()
```

```python
data = data.merge(zip2, on = ['RECORD'], how = 'left')
data.head()
```

```python
data.ZIP = data.ZIP.fillna(data['ZIP_y'])
data.head()
```

```python
del data['ZIP_y']
data.head()
```

```python
data[['ZIP']].info()
```

### Fill in Missing Values by B 

```python
# select record which has null in ZIP
zip = data[data['ZIP'].isnull()][['B','BLOCK','STADDR','ZIP']]
zip['RECORD'] = zip.index + 1 
zip.head()
```

```python
zipfill3 = data_withzip.groupby(['B'])[['ZIP']].agg(lambda x: stats.mode(x)[0][0]).reset_index()
zipfill3.head()
```

```python
zip3 = zip.merge(zipfill3, on = ['B'], how = 'left')
zip3 = zip3[['RECORD','ZIP_y']]
zip3.head()
```

```python
data = data.merge(zip3, on = ['RECORD'], how = 'left')
data.ZIP = data.ZIP.fillna(data['ZIP_y'])
data.head()
```

```python
del data['ZIP_y']
data[['ZIP']].info()
```

```python
data.to_csv('NY property data_zip.csv')
```
